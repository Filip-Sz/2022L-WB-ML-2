# 2. Informacje o danych
Tutaj dokonane będzie ogólne zbadanie danych: ich rozmiar, typy zmiennych i braki.

#### Opis danych:
`school` - student's school (binary: 'GP' - Gabriel Pereira or 'MS' - Mousinho da Silveira)  
`sex` - student's sex (binary: 'F' - female or 'M' - male)  
`age` - student's age (numeric: from 15 to 22)  
`address` - student's home address type (binary: 'U' - urban or 'R' - rural)  
`famsize` - family size (binary: 'LE3' - less or equal to 3 or 'GT3' - greater than 3)  
`Pstatus` - parent's cohabitation status (binary: 'T' - living together or 'A' - apart)  
`Medu` - mother's education (numeric: 0 - none, 1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)  
`Fedu` - father's education (numeric: 0 - none, 1 - primary education (4th grade), 2 – 5th to 9th grade, 3 – secondary education or 4 – higher education)  
`Mjob` - mother's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')  
`Fjob` - father's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')  
`reason` - reason to choose this school (nominal: close to 'home', school 'reputation', 'course' preference or 'other')  
`guardian` - student's guardian (nominal: 'mother', 'father' or 'other')  
`traveltime` - home to school travel time (numeric: 1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour)  
`studytime` - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)  
`failures` - number of past class failures (numeric: n if 1<=n<3, else 4)  
`schoolsup` - extra educational support (binary: yes or no)  
`famsup` - family educational support (binary: yes or no)  
`paid` - extra paid classes within the course subject (Math or Portuguese) (binary: yes or no)  
`activities` - extra-curricular activities (binary: yes or no)  
`nursery` - attended nursery school (binary: yes or no)  
`higher` - wants to take higher education (binary: yes or no)  
`internet` - Internet access at home (binary: yes or no)  
`romantic` - with a romantic relationship (binary: yes or no)  
`famrel` - quality of family relationships (numeric: from 1 - very bad to 5 - excellent)  
`freetime` - free time after school (numeric: from 1 - very low to 5 - very high)  
`goout` - going out with friends (numeric: from 1 - very low to 5 - very high)  
`Dalc` - workday alcohol consumption (numeric: from 1 - very low to 5 - very high)  
`Walc` - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)  
`health` - current health status (numeric: from 1 - very bad to 5 - very good)  
`absences` - number of school absences (numeric: from 0 to 93)  
  
`G1` - first period grade (numeric: from 0 to 20)  
`G2` - second period grade (numeric: from 0 to 20)  
`G3` - final grade (numeric: from 0 to 20, output target)  

`class` - attended course (nominal: 'math' or 'portuguese')

# 3. Korelacje między zmiennymi

```python
plt.figure(figsize=(12,8))
corr = data.corr()
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, annot=True, fmt=".2f")
plt.show()
```

Z powyższej tabelki można odczytać, iż najmocniej ze sobą skorelowane są 3 ostatnie zmienne: `G1`, `G2` i `G3`, czyli końcowe oceny po kolejnych trzech latach nauki.

```python
sns.pairplot(data, x_vars='G1', y_vars=['G2','G3'], height=3)
plt.show()

sns.pairplot(data, x_vars='G2', y_vars='G3', height=3)
plt.show()
```

Oprócz tego mocno skolerowane są ze sobą zmienne `Dalc` i `Walc`, które informują o konsumpcji alkoholu (odpowiednio) w dni robocze i weekendy, ...

```python
sns.pairplot(data, x_vars='Dalc', y_vars='Walc', height=4)
plt.show()
```

... zmienne `Medu` i `Fedu`, mówiące o stopniu wykształcenia (odpowiednio) matki i ojca ucznia...

```python
sns.pairplot(data, x_vars='Medu', y_vars='Fedu', height=4)
plt.show()
```

... oraz zmienne `Walc` (konsumpcja alkoholu w weekendy) i `goout` (wychodzenie ze znajomymi).

```python
sns.pairplot(data, x_vars='Walc', y_vars='goout', height=4)
plt.show()
```

# 4. Rozkłady zmiennych
Tutaj zaprezentowane będą histogramy wybranych zmiennych, które prezentują ich rozkłady.

Rozkład zmiennej `age`:

```python
sns.histplot(data['age'], bins = 20)
plt.show()
```

Najwięcej uczniów jest z grupy wiekowej 16-17 lat.

Rozkład zmiennej `absences`:

```python
sns.histplot(data['absences'],bins = 20)
plt.show()
```

Ogólna frekwencja uczniów jest dobra (rozkład zmiennej "szybko spada"), jedynie pojedyncze osoby mają liczbę nieobecnosći *>50*.

Rozkład zmiennych `G1`, `G2`, `G3`:

```python
data[['G1','G2','G3']].hist(bins = 40, figsize=(18, 9))
plt.show()
```

Rozkład tych zmiennych jest prawie symetryczny, jedynie dla `G2` i `G3` wartość *0.0* częściej się pojawia oraz z każdym rokiem, rozkład coraz bardziej się "spłaszcza".

# 5. Wpływ zmiennych kategorycznych na rozkłady innych

Rozkład wieku z uwzględnieniem relacji romantycznych:

```python
plt.figure(figsize=(8,6))
plt.hist(data[data['romantic']=='yes']['age'], bins=20, alpha=0.5, label="yes")
plt.hist(data[data['romantic']=='no']['age'], bins=20, alpha=0.5, label="no")
plt.xlabel("age", size=14)
plt.ylabel("Count", size=14)
plt.legend(loc='upper right')
plt.title(label='romantic')
plt.show()

```

Rozkład zmiennej wieku osób w związkach jest niższy i bardziej "rozłożony".

Wpływ płci na liczbę nieobecnosći:

```python
plt.figure()
plt.hist(data[data['sex']=='F']['absences'], bins = 20, alpha=0.5, label='female')
plt.hist(data[data['sex']=='M']['absences'], bins = 20, alpha=0.5, label='male')
plt.xlabel("absences", size=14)
plt.ylabel("Count", size=14)
plt.legend(loc='upper right')
plt.title(label='sex')
plt.show()
```

Zmienna reprezentująca liczbę nieobecności ma dla osób płci żeńskiej wartości odstające oraz rozkład tej zmiennej jest wyższy niż dla osób płci męskiej.

Wpływ płci na ilość spożywanego alkoholu (w weekendy):

```python
plt.figure()
plt.hist(data[data['sex']=='F']['Walc'], bins = 20, alpha=0.5, label='female')
plt.hist(data[data['sex']=='M']['Walc'], bins = 20, alpha=0.5, label='male')
plt.xlabel("Walc", size=14)
plt.ylabel("Count", size=14)
plt.legend(loc='upper right')
plt.title(label='sex')
plt.show()
```

Z tego histogramu można wywnioskować, iż większa część osób, która stwierdziła u siebie dużą ilość zpożywanego alkoholu, to osoby płci męskiej.
